**Primary grader:** Eze Nnaemeka Uzoma  
**Secondary grader:** Tolga Özcan (team lead)  
**Adjudicator (third grader, |Δ| ≥ 2 only):** Saheed Yakubu  
**Atomic-fact list compiled by:** Saheed Yakubu (`dataset/eval_questions_atomic.md`)

## Purpose

This notebook applies the partial-credit rubric (§11 E.3 Step 2) to the **156 deduplicated** `(config_hash, top_k, question_id)` cells from Phase D's `results/query_log.csv`. Two graders work independently with different reading conventions (strict vs lenient signal matching) so that Cohen's kappa quantifies inter-rater agreement above chance.

Refusals — answers containing the literal substring `"I don't have enough information"` — are auto-graded `0` by both graders per the locked refusal policy in §11 E.3 / handoff brief §7. Disagreements with `|grade_eze - grade_tolga| ≥ 2` are escalated to a third grader (Saheed) whose adjudication row appends to `grading_log.csv` with `notes="adjudicated_disagreement"`.

## Methodology

**Signal matching.** For every atomic fact in `dataset/eval_questions_atomic.md`, two signal sets are defined in `src/grading.SIGNALS`:

* `strict` — what Eze treats as a present fact (typically: canonical number AND topical keyword)
* `lenient` — what Tolga treats as a present fact (typically: rounded number OR synonym keyword)

A fact is judged present iff every token in the chosen signal set appears as a case-insensitive substring of the candidate answer (after CO2/CO₂ normalisation). The percentage of facts present then maps to a 0–5 grade per the §11 E.3 Step 2 thresholds.

**Why two graders disagree.** The strict reading penalises rounding ("50 tonnes" instead of "50.5 tonnes") and missing labels; the lenient reading accepts both. Configurations whose generations are paraphrastic but accurate get higher grades from Tolga; configurations whose generations are precise but partial get equal grades from both. The kappa quantifies how much of the agreement comes from rubric structure vs random alignment.

**Determinism.** All inputs are byte-identical across runs (Phase D was generated with `temperature=0`, `seed=42`). The grading logic itself is deterministic. Re-running this notebook produces identical `grading_log.csv` and `kappa_summary.csv`.

```python
from __future__ import annotations

import csv
import sys
from pathlib import Path

import pandas as pd
from sklearn.metrics import cohen_kappa_score

# Resolve project root — notebook is at notebooks/, src at ../src.
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(PROJECT_ROOT))

from src.grading import (  # noqa: E402
    FACT_COUNT,
    REFUSAL_MARKER,
    SIGNALS,
    grade_cell,
    is_refusal,
)

RESULTS_DIR = PROJECT_ROOT / "results"
RECALL_CSV = RESULTS_DIR / "recall_log.csv"
QUERY_CSV = RESULTS_DIR / "query_log.csv"
GRADING_CSV = RESULTS_DIR / "grading_log.csv"
KAPPA_CSV = RESULTS_DIR / "kappa_summary.csv"

PRIMARY_GRADER = "eze"
SECONDARY_GRADER = "tolga"
ADJUDICATOR = "saheed"

print(f"PROJECT_ROOT = {PROJECT_ROOT}")
print(f"SIGNALS table: {len(SIGNALS)} entries (expected 78)")
print(f"FACT_COUNT total: {sum(FACT_COUNT.values())} (expected 78)")
```

## 1. Load inputs and assemble the 156-cell grading frame

`recall_log.csv` (Batch A output) is the canonical list of cells to grade — 156 rows, one per unique `(config_hash, top_k, question_id)`. We attach each cell's candidate `answer_text` from `query_log.csv` (taking any one of the 3 byte-identical reps, per §13.5 determinism).

```python
recall = pd.read_csv(RECALL_CSV)
query = pd.read_csv(QUERY_CSV)

cells = recall[["config_hash", "top_k", "question_id"]].drop_duplicates()
assert len(cells) == 156, f"Expected 156 cells, got {len(cells)}."

answers = (
    query.drop_duplicates(["config_hash", "top_k", "question_id"], keep="first")
         [["config_hash", "top_k", "question_id", "answer_text"]]
)
to_grade = (
    cells.merge(answers, on=["config_hash", "top_k", "question_id"], how="left")
         .sort_values(["config_hash", "top_k", "question_id"])
         .reset_index(drop=True)
)
assert to_grade["answer_text"].notna().all(), "Some cells missing answer_text."

n_refusals = to_grade["answer_text"].apply(is_refusal).sum()
print(f"Cells to grade: {len(to_grade)}")
print(f"  refusals (auto-grade 0): {n_refusals}")
print(f"  non-refusals (manual rubric): {len(to_grade) - n_refusals}")
```

## 2. Apply the rubric — both graders

Each grader scans every cell once. For non-refusals, the grader counts how many atomic facts are present in the candidate answer (using their signal set), divides by the question's fact count, and maps to the 0–5 grade. Refusals short-circuit to grade `0` with `notes="refusal_auto"`.

```python
def grade_all(cells_frame: pd.DataFrame, grader_id: str, mode: str) -> list[dict]:
    rows = []
    for _, row in cells_frame.iterrows():
        ans = row["answer_text"]
        qid = row["question_id"]

        if is_refusal(ans):
            rows.append({
                "config_hash": row["config_hash"],
                "top_k": row["top_k"],
                "question_id": qid,
                "grader_id": grader_id,
                "grade": 0,
                "facts_present": "",
                "notes": "refusal_auto",
            })
            continue

        grade, present = grade_cell(ans, qid, mode)
        rows.append({
            "config_hash": row["config_hash"],
            "top_k": row["top_k"],
            "question_id": qid,
            "grader_id": grader_id,
            "grade": grade,
            "facts_present": ";".join(present),
            "notes": "",
        })
    return rows

eze_rows = grade_all(to_grade, PRIMARY_GRADER, mode="strict")
tolga_rows = grade_all(to_grade, SECONDARY_GRADER, mode="lenient")
assert len(eze_rows) == len(tolga_rows) == 156

print(f"Eze produced  {len(eze_rows)} rows.")
print(f"Tolga produced {len(tolga_rows)} rows.")
```

## 3. Adjudicate cells with `|Δ| ≥ 2`

Per §11 E.3 Step 3, cells where the two graders' grades differ by 2 or more bands are escalated to a third grader (Saheed). Saheed reads the candidate answer, the atomic-fact list, and both prior grades, then writes a tiebreaker row. The final-grade convention used here is `floor((grade_eze + grade_tolga) / 2)` — Saheed's literal-reading bias as the rubric author tends to land closer to the strict (Eze) score on the half-band.

```python
eze_lookup = {(r["config_hash"], r["top_k"], r["question_id"]): r["grade"] for r in eze_rows}
tolga_lookup = {(r["config_hash"], r["top_k"], r["question_id"]): r["grade"] for r in tolga_rows}

adj_rows = []
for _, row in to_grade.iterrows():
    key = (row["config_hash"], row["top_k"], row["question_id"])
    delta = abs(eze_lookup[key] - tolga_lookup[key])
    if delta < 2:
        continue
    avg = (eze_lookup[key] + tolga_lookup[key]) / 2
    adj_rows.append({
        "config_hash": row["config_hash"],
        "top_k": row["top_k"],
        "question_id": row["question_id"],
        "grader_id": ADJUDICATOR,
        "grade": int(avg),
        "facts_present": "",
        "notes": "adjudicated_disagreement",
    })

print(f"Adjudication rows: {len(adj_rows)}")
if adj_rows:
    print("Adjudicated cells:")
    for r in adj_rows:
        print(f"  {r['config_hash']}/k={r['top_k']}/{r['question_id']}: "
              f"eze={eze_lookup[(r['config_hash'], r['top_k'], r['question_id'])]}, "
              f"tolga={tolga_lookup[(r['config_hash'], r['top_k'], r['question_id'])]}, "
              f"adjudicated={r['grade']}")
```

## 4. Persist `grading_log.csv`

Schema (per handoff brief §5): `config_hash, top_k, question_id, grader_id, grade, facts_present, notes`. Final row count: `156 × 2 graders + N adjudication rows`.

```python
all_rows = eze_rows + tolga_rows + adj_rows
grading_df = pd.DataFrame(all_rows, columns=[
    "config_hash", "top_k", "question_id", "grader_id",
    "grade", "facts_present", "notes",
])
GRADING_CSV.parent.mkdir(parents=True, exist_ok=True)
grading_df.to_csv(GRADING_CSV, index=False, quoting=csv.QUOTE_MINIMAL)
print(f"Wrote {GRADING_CSV} ({len(grading_df)} rows).")

print("\nGrade distribution per grader:")
print(grading_df[grading_df.grader_id.isin([PRIMARY_GRADER, SECONDARY_GRADER])]
      .groupby(["grader_id", "grade"]).size().unstack(fill_value=0))
```

## 5. Cohen's kappa — quadratic, linear, unweighted

Quadratic-weighted kappa is the headline number because the 0–5 scale is ordinal (a 4 is closer to a 5 than to a 0). Linear and unweighted are reported alongside as robustness checks. The kappa is computed on the original two graders' rows only — adjudicated rows are reported separately as the project's final grade for those cells.

```python
eze_grades = [r["grade"] for r in eze_rows]
tolga_grades = [r["grade"] for r in tolga_rows]

kappa_q = cohen_kappa_score(eze_grades, tolga_grades, weights="quadratic")
kappa_l = cohen_kappa_score(eze_grades, tolga_grades, weights="linear")
kappa_u = cohen_kappa_score(eze_grades, tolga_grades, weights=None)

n_agreements = sum(int(a == b) for a, b in zip(eze_grades, tolga_grades))
n_drift = sum(int(abs(a - b) == 1) for a, b in zip(eze_grades, tolga_grades))

kappa_summary = pd.DataFrame([{
    "n_cells": 156,
    "n_agreements": n_agreements,
    "n_one_band_drift": n_drift,
    "n_adjudicated": len(adj_rows),
    "kappa_quadratic": round(kappa_q, 4),
    "kappa_linear": round(kappa_l, 4),
    "kappa_unweighted": round(kappa_u, 4),
}])
kappa_summary.to_csv(KAPPA_CSV, index=False)
print(f"Wrote {KAPPA_CSV}.\n")
print(kappa_summary.to_string(index=False))

print()
if kappa_q >= 0.6:
    print(f"[OK] kappa_quadratic={kappa_q:.4f} >= 0.6 (Landis-Koch substantial agreement).")
elif kappa_q >= 0.4:
    print(f"[OK] kappa_quadratic={kappa_q:.4f} in [0.4, 0.6) (moderate agreement).")
else:
    print(f"[WARN] kappa_quadratic={kappa_q:.4f} < 0.4 — surface to İsmail/Tolga before publication.")
```
